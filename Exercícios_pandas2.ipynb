```python
#Tabela com 10 clientes com maior saldo (balance)
def min_max(data):
    style = data.copy()
    max_val = style.max()
    min_val = style.min()
    style = ['background-color: yellow' if v == max_val else 'background-color: lightcoral' if v == min_val else '' for v in style]
    return style
    
top_10 = bank[['age', 'job', 'marital', 'education', 'balance', 'default', 'loan', 'duration']].sort_values(by='balance', ascending=False).head(10)
top_10_format = top_10.style.apply(min_max, subset=['duration'])
display(top_10_format)
```

```python
#Dataset bikes

#Importando produtos
bike_products = pd.read_csv('data/bike_shop/products.csv')
bike_products.head()
```

```python
#Importando marcas
bike_brands = pd.read_csv('data/bike_shop/brands.csv')
bike_brands.head()
```

```python
#M√©dia por brand_id
avg = bike_products.groupby('brand_id')['list_price'].mean().reset_index()

#Unindo datasets
media_marcas = avg.merge(bike_brands, on ='brand_id', how='left')

#Ordenando valores
media_marcas_ord = media_marcas.sort_values(by= 'list_price', ascending=False)

media_marcas_ord
```

```python
#Criando faixas de pre√ßo
media_marcas_ord['faixa_de_preco'] = pd.qcut(media_marcas_ord['list_price'], q=[0, 1/3, 2/3, 1] , labels =['Barato', 'M√©dio', 'Caro'])

media_marcas_ord
```

```python
#Vendas por faixa de preco
```

```python
#Importando datasets
order_items = pd.read_csv('data/bike_shop/order_items.csv')
orders = pd.read_csv('data/bike_shop/orders.csv')
```

```python
orders.shape
```

```python
#unindo datasets
df_unificado = bike_brands.merge(bike_products, on='brand_id', how='left').merge(order_items, on='product_id', how='left').merge(orders, on='order_id', how ='left')
df_unificado_marca = df_unificado.merge(media_marcas_ord[['brand_name', 'faixa_de_preco']], on='brand_name', how = 'left')
```

```python
#Agrupando pelo total de vendas
vendas_por_marca = df_unificado_marca.groupby('brand_name').agg(
    qntd_vendas=('quantity', 'sum'),
    faixa_de_preco=('faixa_de_preco', 'first')
).sort_values(by='qntd_vendas', ascending=False).reset_index()

vendas_por_marca
```

A quantidade de produtos vendidos por faixa de pre√ßo pode variar bastante, o que indica que existem outros fatores que podem influenciar mais nas vendas do que o pre√ßo. 

```python
#Vendas x Custo por Estado
```

```python
#Carregando datasets necess√°rios
stores = pd.read_csv('data/bike_shop/stores.csv')
stocks = pd.read_csv('data/bike_shop/stocks.csv')
stores.head(10)
```

```python
#Agrupando datasets
vendas_estado = orders[['order_id', 'store_id']].merge(stores[['store_id', 'state', 'store_name']], on ='store_id', how='left')
vendas_estado
```

```python
#Agrupando o estoque por estado
estoque = (
    stocks
    .merge(stores[['store_id', 'state']], on='store_id', how='left')
    .groupby('state', as_index=False)['quantity'].sum()
    .rename(columns={'quantity': 'estoque_total'})
)

estoque
```

```python
#Agrupando as vendas por estado
vendas_por_estado = vendas_estado.groupby('state', as_index=False)['order_id'].count().rename(columns={'order_id':'vendas'})

vendas_por_estado
```

```python
#Unindo as tabelas
estado_resumo = vendas_por_estado.merge(estoque, on='state', how='left')

estado_resumo
```

NY √© o estado com mais vendas disparadamente, e ao mesmo tempo possui o menor estoque dispon√≠vel. 

CA e TX apresentam valores de venda consideravelmente mais baixos, com destaque para o Texas. Em ambos os estado, temos valores de estoque em patamares semelhantes.
